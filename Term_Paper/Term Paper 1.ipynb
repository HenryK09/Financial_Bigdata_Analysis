```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
print(Y)
```

```python
X_train = df[["mpg" , "cylinders" , "cubicinches" , "hp" , "weightlbs" , "time-to-60", "year"]].values
X_train = np.array(X_train, dtype = np.float32)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)

Y_train = Y
one_hot_dummy = tf.one_hot(Y, depth = 3).numpy()
Y_train = one_hot_dummy

print(X_train, 'shape = ', X_train.shape)
print(Y_train, 'shape = ', Y_train.shape)
```

```python
tf.random.set_seed(2021)
W = tf.Variable(tf.random.normal([X_train.shape[1],Y_train.shape[1]], mean=0.0))
b = tf.Variable(tf.random.normal([Y_train.shape[1]], mean=0.0))
print('# Weights: \n', W.numpy(), '\n\n# Bias: \n', b.numpy())
```

```python
def softmax(X):
    z = tf.matmul(X, W) + b
    sm = tf.nn.softmax(z)
    return sm

def cost_fn(Y_pred, Y_true):
    cost = tf.reduce_mean(-tf.reduce_sum(Y_true * tf.math.log(Y_pred), axis=1))
    return cost

learning_rate = 0.01
optimizer = tf.optimizers.SGD(learning_rate)
```

```python
for step in range(10001):
        
    with tf.GradientTape() as g:
        pred = softmax(X_train)
        cost = cost_fn(pred, Y_train)
            
        gradients = g.gradient(cost, [W, b])
    
    optimizer.apply_gradients(zip(gradients,[W, b]))
    
    if step % 1000 == 0:
            
        print("step =", '%04d' % (step+1), "cost = {}, W = {}, b = {}".format(cost, W.numpy(), b.numpy()))

print('*'*50)
print(f'y = {W.numpy()}x + {b.numpy()}')
```

```python
predicted = tf.argmax(softmax(X_train), axis = 1)
real      = tf.argmax(Y_train, axis = 1)

def accuracy(hypo, label):
    accuracyd = tf.reduce_mean(tf.cast(tf.equal(hypo, label), dtype=tf.float32))
    return accuracyd

acc = accuracy(predicted, real).numpy()
print("Accuracy: %s" % (acc*100))
```
